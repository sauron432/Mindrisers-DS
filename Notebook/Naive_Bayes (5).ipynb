Probability

Independent Events
- Rolling a dice {1,2,3,4,5,6} P(1)--> 1/6, P(2)-->1/6

Dependent Events
- Bag Of Marbles (3 oranges and 2 yellow). Probability of first removing a orange marble and then a yellow marble? P(o) = 3 /5 P(y) = P(y/o) = 2/4

P(o and y) = P(o) * P(y/o)--> Conditional Probability

P(A and B) = P(A) * P(B/A)

- P(A and B) = P(B and A)
- P(A) * P(B/A) = P(B) * P(A/B)
- P(A/B) = (P(A) * P(B/A)) / P(B)

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, y = load_iris(return_X_y = True, as_frame = True)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
gnb = GaussianNB()
bnb = BernoulliNB()
mnb = MultinomialNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
bnb.fit(X_train, y_train)
```

```python
mnb.fit(X_train, y_train)
```

```python
y_pred_gnb = gnb.predict(X_test)
y_pred_bnb = bnb.predict(X_test)
y_pred_mnb = mnb.predict(X_test)
```

```python
y_pred_gnb
```

```python
y_pred_bnb
```

```python
y_pred_mnb
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred_bnb))
print(classification_report(y_test, y_pred_bnb))
print(accuracy_score(y_test, y_pred_bnb))
```

```python
print(confusion_matrix(y_test, y_pred_mnb))
print(classification_report(y_test, y_pred_mnb))
print(accuracy_score(y_test, y_pred_mnb))
```

```python
print(confusion_matrix(y_test, y_pred_gnb))
print(classification_report(y_test, y_pred_gnb))
print(accuracy_score(y_test, y_pred_gnb))
```
