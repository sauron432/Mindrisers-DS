Now we will check the relationship between two numerical data.
Number of cells and Range

```python
#Scatter plot
plt.scatter(newdf['battery_capacity_kWh'],newdf['range_km'], color='orange')
#sorted_battery_capacity = newdf.sort_values('battery_capacity_kWh')
#plt.plot(sorted_battery_capacity['battery_capacity_kWh'] , newdf['range_km'])
plt.xlabel('Range')
plt.ylabel('No of cells')
plt.show()
```

From the scatter plot above, we can say that number of cells and range have positive relation.

```python
awd_top_speed = newdf[newdf['drivetrain'] == 'AWD']['top_speed_kmh']
rwd_top_speed = newdf[newdf['drivetrain'] == 'RWD']['top_speed_kmh']
fwd_top_speed = newdf[newdf['drivetrain'] == 'FWD']['top_speed_kmh']
awd_range = newdf[newdf['drivetrain'] == 'AWD']['range_km']
rwd_range = newdf[newdf['drivetrain'] == 'RWD']['range_km']
fwd_range = newdf[newdf['drivetrain'] == 'FWD']['range_km']
```

```python
plt.boxplot([awd_top_speed, fwd_top_speed, rwd_top_speed], tick_labels = ['AWD','FWD','RWD'])
plt.xlabel('Drivetrain')
plt.ylabel('Top speed (km/h)')
plt.show()
```

```python
plt.boxplot([awd_range, fwd_range, rwd_range], tick_labels = ['AWD','FWD','RWD'])
plt.xlabel('Drivetrain')
plt.ylabel('Range')
plt.show()
```
