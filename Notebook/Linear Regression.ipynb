```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
sns.set()
```

```python
df = pd.read_csv('1.01.Simple-linear-regression.csv')
df.head()
```

```python
df.describe()
```

```python
y = df['GPA']
x1 = df['SAT']
plt.scatter(x1,y)
plt.grid(True,linestyle='--', color='green', alpha=0.2)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1,yhat, c='orange')
# plt.grid(True,linestyle='--', color='green', alpha=0.2)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```
